```python
import subprocess
import threading
import asyncio
import time

def start_ollama():
  process = subprocess.Popen("ollama serve", shell=True)

def run_ollama_async():
  thread = threading.Thread(target=start_ollama)
  thread.start()

run_ollama_async()

time.sleep(10)  # 等待 10 秒

!ollama run llama3:8b
```
